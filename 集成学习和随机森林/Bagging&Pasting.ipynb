每个预测器使用的算法相同，但是在不同的训练集随机子集上进行训练。采样时如果将样本放回，这种方法叫做bagging（自举汇聚法/bootstrap aggregating）；如果样本不放回，这种方法则叫做pasting

每个预测器的单独的偏差都高于在原始训练集上训练的偏差，但是通过聚合，同时降低了偏差和方差。

总的来说，与直接在原始训练集上训练的单个预测器相比，集成的偏差相近但是方差更低。

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
import numpy as np

bag_clf=BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500,
    max_samples=100, bootstrap=True, n_jobs=-1
)
iris=datasets.load_iris()
X_train=iris["data"][:, (2,3)]
y_train=(iris["target"]==2).astype(np.float64)
bag_clf.fit(X_train, y_train)
y_pred=bag_clf.predict(X_train)

#包外评估
bag_clf=BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500,
    bootstrap=True, n_jobs=-1, oob_score=True
)
bag_clf.fit(X_train, y_train)
bag_clf.oob_score_

from sklearn.metrics import accuracy_score
y_pred=bag_clf.predict(X_train)
accuracy_score(y_train, y_pred)

bag_clf.oob_decision_function_
```
