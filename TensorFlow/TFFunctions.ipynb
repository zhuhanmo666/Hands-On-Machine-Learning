```python
import tensorflow as tf
def cube(x):
    return x**3
tf_cube = tf.function(cube)
@tf.function
def tf_cube(x):
    return x**3
```

**TF函数规则**:
1. 如果调用任何外部库，包括NumPy甚至标准库，此调用将仅在跟踪过程中运行。它不会成为图表的一部分。
2. 如果你定义了一个返回np.random.rand()的TF函数f(x)，则仅在跟踪该函数时才会生成随机数，因此f(t.constant(2.))和f(t .constant(3.))将返回相同的随机数，但f(t.constant([2.，3.]))将返回不同的随机数
3. 如果你的非TensorFlow代码有副作用（例如记录某些内容或新Python计数器），那么你不应期望每次调用TF函数时都会发生这些副作用，因为它们只会在跟踪该函数时发生
4. 你可以在t .py_function()操作中包装任何Python代码，但这样做会降低性能，因为TensorFlow无法对此代码进行任何图优化,也会降低可移植性
5. 你可以调用其他Python函数或TF函数，但它们应遵循相同的规则，因为TensorFlow会在计算图中捕获它们的操作
6. 如果该函数创建了一个TensorFlow变量，则必须在第一次调用时这样做否则你会得到一个异常
7. Python函数的源代码应可用于TensorFlow
8. TensorFlow只能捕获在张量或数据集上迭代的or循环。使用`for i in tf.range(x)`，而不是使用`for i in range(x)`
9. ，应尽可能使用向量化实现，而不是使用循环
