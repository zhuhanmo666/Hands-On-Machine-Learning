```python
import numpy as np

env_display = gym.make("CartPole-v1", render_mode="human")
env_array = gym.make("CartPole-v1", render_mode="rgb_array")

obs_display = env_display.reset()
obs_array = env_array.reset()

for i in range(50):
    # 显示窗口
    env_display.render()
    action = env_display.action_space.sample()
    
    # 对两个环境执行相同的操作
    obs_display, reward, terminated, truncated, info = env_display.step(action)
    obs_array, _, _, _, _ = env_array.step(action)
    
    if terminated or truncated:
        obs_display = env_display.reset()
        obs_array = env_array.reset()

env_display.close()
env_array.close()
```

```python
def basic_policy(obs):
    angle = obs[2]  # obs应该是包含4个元素的数组 [position, velocity, angle, angular_velocity]
    return 0 if angle < 0 else 1

env = gym.make("CartPole-v1", render_mode="human")

totals = []
for episode in range(500):
    episode_rewards = 0
    obs, info = env.reset()  # 现在reset返回(observation, info)元组
    for step in range(200):
        action = basic_policy(obs)
        obs, reward, terminated, truncated, info = env.step(action)  # 5个返回值
        episode_rewards += reward
        done = terminated or truncated  # 统一的结束条件
        if done:
            break
    totals.append(episode_rewards)

env.close()
print(np.mean(totals), np.std(totals), np.min(totals), np.max(totals))
```
